Handling nulls

```python
#replace nulls with mean values and remove white spaces
mean_tenure = data['Tenure'].mean()
data['Tenure'].fillna(mean_tenure, inplace = True)

mean_warehouse_to_home = data['WarehouseToHome'].mean()
data['WarehouseToHome'].fillna(mean_warehouse_to_home, inplace = True)

mean_hours_on_app = data['HourSpendOnApp'].mean()
data['HourSpendOnApp'].fillna(mean_hours_on_app, inplace = True)

mean_hikes_from_last_year = data['OrderAmountHikeFromlastYear'].mean()
data['OrderAmountHikeFromlastYear'].fillna(mean_hikes_from_last_year, inplace = True)

mean_coupon_used = data['CouponUsed'].mean()
data['CouponUsed'].fillna(mean_coupon_used, inplace = True)

mean_order_count = data['OrderCount'].mean()
data['OrderCount'].fillna(mean_order_count, inplace = True)

mean_days_since_last_order = data['DaySinceLastOrder'].mean()
data['DaySinceLastOrder'].fillna(mean_days_since_last_order, inplace = True)

#checking if there are white spaces
whitespace_check = data.apply(lambda x: x.astype(str).str.isspace().any())
whitespace_check
```

Checking consistency and accuracy

```python
#Checking if all strings are correct and meaningful
print(data['PreferredLoginDevice'].unique())
print(data['PreferredPaymentMode'].unique())
print(data['Gender'].unique())
print(data['PreferedOrderCat'].unique())
print(data['MaritalStatus'].unique())
```

```python
#In prefered login device Mobile Phone is same as Phone
for index, items in enumerate(data['PreferredLoginDevice']): 
    if items == 'Phone': 
        data['PreferredLoginDevice'][index] = items.replace(items, 'Mobile Phone')
        
#in preferred payment modes, CC is same as credit card and COD & Cash on Delivery is same as Cash On Delivery.
for index, items in enumerate(data['PreferredPaymentMode']): 
    if items == 'CC': 
        data['PreferredPaymentMode'][index] = items.replace(items, 'Credit Card')
        
    elif items == 'COD' or items == 'Cash on Delivery': 
        data['PreferredPaymentMode'][index] = items.replace(items, 'Cash On Delivery')
        
    elif items == 'UPI':
        data['PreferredPaymentMode'][index] = items.replace(items, 'Unified Payment Interface')
        
#In prefered order cat Mobile is same as Mobile Phone
for index, items in enumerate(data['PreferedOrderCat']): 
    if items == 'Mobile': 
        data['PreferedOrderCat'][index] = items.replace(items, 'Mobile Phone')
```

```python
#Checking fixes
print(data['PreferredLoginDevice'].unique())
print(data['PreferredPaymentMode'].unique())
print(data['Gender'].unique())
print(data['PreferedOrderCat'].unique())
print(data['MaritalStatus'].unique())
```

Detecting Outliers

```python
data.describe()
```

```python
#I'll use z-score (threshold = 3) to find outliers
numeric_cols = data.select_dtypes(include= ['float64', 'int64']).columns

#function to detect outliers 
def detect_outliers_zscore(data, threshold = 3): 
    outliers = pd.DataFrame()
    for col in numeric_cols: 
        z_scores = np.abs((data[col] - data[col].mean())/ data[col].std())
        outliers[col] = data[col][z_scores > threshold]
    return outliers

#detecting outliers
outliers = detect_outliers_zscore(data)
```

```python
outliers.info()
```

There are 4 outliers in 'Tenure column'- a column that shows how long a customer has been with the business.Considering that the e-commerce business is likely to have loyal customers for a longer time compared to others; this is a genuine outlier. I decide to keep them since they might provide valuable insights
